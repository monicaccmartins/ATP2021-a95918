```python
import random

def menu():
    print("""Bem vindo! Este programa tem várias funcionalidades. O que queres fazer?
    (1) - Criar fração
    (2) - Simplificar fração
    (3) - Somar frações
    (4) - Subtrair frações
    (5) - Multiplicar frações
    (6) - Dividir frações
    (7) - Gerar lista de n frações
    (8) - Somar lista
    (9) - Maior fração de uma lista
    (0) - Sair
     """)
        
def criarFracao(num, den):
    return (num, den)

def verFracao(f):
    return (str(f[0])+"/"+str(f[1]))
    
def mdcomum(a,b):
    if a%b==0:
        return b
    elif b%a==0:
        return a
    else:
        if a>b:
            return mdcomum(b, a%b)
        else:
            return mdcomum(a, b%a)
        
def simplificarFracao(f):
    num, denom = f
    a=mdcomum(num,denom)
    return (num/a,denom/a)

def somaFrac(f1,f2):
    n1,d1=f1
    n2,d2=f2
    F=criarFracao(n1*d2+n2*d1,d1*d2)
    return simplificarFracao(F)

def subtraiFrac(f1,f2):
    n1,d1=f1
    n2,d2=f2
    F=criarFracao(n1*d2-n2*d1,d1*d2)
    return simplificarFracao(F)

def multiplica(f1,f2):
    n1,d1=f1
    n2,d2=f2
    F=criarFracao(n1*n2,d1*d2)
    return simplificarFracao(F)

def divideFrac(f1,f2):
    n1,d1=f1
    n2,d2=f2
    F=criarFracao(n1*d2,d1*n2)
    return simplificarFracao(F)

def criarListaA():
    N=int(input("Quantos números queres na tua lista? "))
    print(" ")
    L=[]
    LV=[]      #é guardada uma lista com as fraçoes com virgulas para mais tarde conseguir usar esta lista na somaFrac()
    for i in range(N):
        LA=()
        for j in range(2):
            if j==0:
                num=(random.randint(1,1000))
            else:
                denom=(random.randint(0,1000))
        LA=(num,denom)
        Lv=(simplificarFracao(LA))
        LV.append(Lv)
        Ls=verFracao(simplificarFracao(LA))
        L.append(Ls)
    return (L,LV)

def comparar(L):
    max=L[0]
    for i in range(1,len(L)):
        n1,d1=max
        n2,d2=L[i]
        if (n1*d2)<(n2*d1):
            max=L[i]            
            
    return max
        
        

Lsimp=[]    #para os resultados anteriores serem apagados
Lvirg=[]
op=1
while op!=0:
    print(" ")
    menu()
    op=int(input("Opção: "))
    print(" ")
    if op==1:
        num=int(input("Numerador: "))
        den=int(input("Denominador: "))
        print(verFracao(criarFracao(num,den)))
        
    
    elif op==2: 
        num=int(input("Numerador: "))
        den=int(input("Denominador: "))
        print(verFracao(simplificarFracao(criarFracao(num,den))))
        
    
    elif op==3:
        n=int(input("Quantas fraçoes queres somar?"))
        L3=[]
        for i in range(n):
            num=int(input("Numerador: "))
            den=int(input("Denominador: "))
            L1=[num,den]
            L3.append(L1)
        soma=(0,1)
        for elem in L3:
            soma=somaFrac(soma,elem)
        print(verFracao(simplificarFracao(soma)))
    
    elif op==4:
        L4=[]
        for i in range(2):
            num=int(input("Numerador: "))
            den=int(input("Denominador: "))
            L1=[num,den]
            L4.append(L4)
        print(verFracao(simplificarFracao(subtraiFrac(L4[0],L4[1]))))
        
        
    elif op==5:
        n=int(input("Quantas fraçoes queres multiplicar?"))
        L5=[]
        for i in range(n):
            num=int(input("Numerador: "))
            den=int(input("Denominador: "))
            L1=[num,den]
            L5.append(L1)
        mult=(1,1)
        for elem in L5:
            mult=multiplica(mult,elem)
        print(verFracao(simplificarFracao(mult)))
        
        
    elif op==6:
        L6=[]
        num1=int(input("Numerador do dividendo: "))
        den1=int(input("Denominador do dividendo: "))
        L1=[num1,den1]
        num2=int(input("Numerador do divisor: "))
        den2=int(input("Denominador do divisor: "))
        L2=[num2,den2]
        L6.append(L1)
        L6.append(L2)
        print(verFracao(simplificarFracao(divideFrac(L6[0],L6[1]))))
        print(" ")
    
    elif op==7:
        Lsimp,Lvirg=criarListaA()
        print(Lsimp)
        
    elif op==8:
        print('A última lista que inseriste foi: '+ str(Lsimp))
        somat=(0,1)
        for i in range(len(Lvirg)):
            somat=somaFrac(somat,Lvirg[i])
        print('A sua soma é: ' + str(verFracao(simplificarFracao(somat))))
    
    elif op==9:
        if len(Lvirg)==0:
            print("Ainda não adicionou elementos à lista!")
        else:
            print(comparar(Lvirg))
    
    elif op==0:
        print("Até à proxima!")
        
    else:
        print("Opção inválida!")
```
