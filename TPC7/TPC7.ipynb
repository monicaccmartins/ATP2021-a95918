```python
import io
import matplotlib.pyplot as plt


# EXERCICIO 1

def getAluno(texto):
    textoLimpo=texto.replace("\"", "")
    campos=textoLimpo.split(",")            
    aluno=[]
    aluno.append(campos[0])
    aluno.append(campos[1])
    aluno.append(campos[2])
    aluno.append(campos[3:])
    lista=[]
    for s in aluno[3]:
        lista.append(int(s))
    aluno[3]=lista
    return aluno

def lerDataset(fnome):
    f=io.open(fnome, encoding="utf-8")
    bd=[]
    f.readline()
    for linha in f:
        bd.append(getAluno(linha))
    return bd


# EXERCICIO 2

def chaveOrd(a):
    return a[1]

def listarDataset(bd):
    bd.sort(key=chaveOrd)
    print("id |        nome        curso     media")
    print("---------------------------------------------")
    for a in bd:
        print(a[0] + " | " + a[1] + " | " + a[2] + " | " + str(sum(a[3])/4))


# EXERCICIO 3

def media(l):
    soma=0
    for i in range(0,len(l)):
        soma=soma+l[i]
    return soma/4


def consultarDataset(bd, id):
    for i in range(0,len(bd)):
        if bd[i][0]==id:
            med=media(bd[i][3])
            return print("INFO :: " + str(bd[i]) + "\nMEDIA :: " + str(med))
        else:
            if i==len(bd)-1:
                return print('Não existe nenhum aluno com esse id!')


# EXERCICIO 4

def chaveOrdena(a):
    return media(a[3])

def top10(bd):                                   
    bd.sort(key=chaveOrdena, reverse= True)           #vamos criar nova lista ordenada pelas medias e imprimir os 10 elementos da lista
    novaBD=[]                                         #criamos nova lista
    for aluno in range(0,10):                         #vamos percorrer os primeiros 10 alunos da bd 
        novoAluno=[]                             
        for parametro in range(0,3):                  #copiamos o id, nome e curso do aluno que estamos a analisar para a nova lista
            novoAluno.append(bd[aluno][parametro])  
        novoAluno.append(media(bd[aluno][3]))         #no fim, adicionamos a media do aluno à lista
        novaBD.append(novoAluno)
    return novaBD
    


# EXERCICIO 5

def distribPorCurso(bd):
    distribuicao ={}
    for a in bd:
        if a[2] in distribuicao.keys():
            distribuicao[a[2]]= distribuicao[a[2]] + 1
        else:
            distribuicao[a[2]]=1
    return distribuicao


# EXERCICIO 6

def distribPorMedia(bd):
    distrib={}
    for a in bd:
        if ('M=' + str(round(media(a[3])))) in distrib.keys(): 
            distrib['M=' + str(round(media(a[3])))]= distrib['M=' + str(round(media(a[3])))]+1
        else:
            distrib['M=' + str(round(media(a[3])))]=1
    return(distrib)


# EXERCICIO 7

def plotDistribPorCurso(bd):
    c=distribPorCurso(bd)
    x=c.keys()
    y=[]
    for i in x:
        y.append(c[i])
    plt.bar(x, y)
    plt.show()
    

# EXERCICIO 8

def plotDistribPorMedia(bd):
   cm=distribPorMedia(bd)
   x=cm.keys()
   y=[]
   for i in x:
        y.append(cm[i])
   plt.bar(x, y)
   plt.show()





BD = []
BD = lerDataset("alunos.csv")  
print("EXERCICIO 1: ")
print("\n")
print(BD)
print("\n")
print("EXERCICIO 2: ")
print("\n")
listarDataset(BD)
print("\n")
print("EXERCICIO 3: ")
print("\n")
consultarDataset(BD, "a95")
print("\n")
print("EXERCICIO 4: ")
print("\n")
for linha in top10(BD):
    print(linha)
print("\n")
print("EXERCICIO 5: ")
print("\n")
print(distribPorCurso(BD))
print("\n")
print("EXERCICIO 6: ")
print("\n")
print(distribPorMedia(BD))
print("\n")
print("EXERCICIO 7: ")
print("\n")
plotDistribPorCurso(BD)
print("\n")
print("EXERCICIO 8: ")
print("\n")
plotDistribPorMedia(BD)    


```
